```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import math
```

```python
df = pd.read_csv('../crawldata/hcm.csv')
df.head(50)
```

```python
from pandas import DataFrame
x_train = DataFrame(df,columns = ['bed','bath','area'])
x_train.head(5)
```

```python
y_train = df.budget
y_train.head(5)
```

```python
data_train =DataFrame(df,columns = ['bed','bath','area','budget'])
data_train
```

```python
reg = linear_model.LinearRegression().fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
test_df = pd.read_csv("../crawldata/hcmtest.csv")
test_df.head(3)
```

```python
x_test = DataFrame(test_df,columns = ['bed','bath','area'])
y_test = DataFrame(test_df,columns = ['budget'])
x_test
```

```python
reg.score(x_test,y_test)
```

```python
demo_df = pd.read_csv('../crawldata/hcmtest.csv')
demo = DataFrame(demo_df,columns = ['bed','bath','area'])
demo.head(3)
```

```python
demo['budget_guess']= reg.predict(demo)
```

```python
demo[['budget','local_street','local_ward','local_district','local_city']] = demo_df[['budget','local_street','local_ward','local_district','local_city']]
demo.head(20)
```
